```python
import pandas as pd
import os

# Load the dataset (update the path if needed)
df = pd.read_csv("car_details.csv")

# Extract brand name (first word of the 'name' column)
df['brand'] = df['name'].str.split().str[0]

# Create a folder to save the brand files
os.makedirs("brand_files", exist_ok=True)

# Loop through each brand and save a separate CSV
for brand in df['brand'].unique():
    brand_df = df[df['brand'] == brand].drop(columns=['brand'])
    filename = f"brand_files/{brand}.csv"
    brand_df.to_csv(filename, index=False)
    print(f"Saved: {filename} ({len(brand_df)} rows)")

print("\nDone! All brand files saved in the 'brand_files' folder.")
```

```python
import matplotlib.pyplot as plt

brands = ['Maruti', 'Hyundai', 'Mahindra', 'Tata', 'Honda']
counts = [1280, 821, 365, 361, 252]

plt.figure(figsize=(10, 6))
plt.bar(brands, counts, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])

plt.title('Top 5 Car Brands by Number of Listings', fontsize=16)
plt.xlabel('Car Brand', fontsize=13)
plt.ylabel('Number of Cars', fontsize=13)

for i, count in enumerate(counts):
    plt.text(i, count + 10, str(count), ha='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```
